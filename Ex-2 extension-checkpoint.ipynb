# pattern in time series data (beehive method)

###### follow this link for the site: https://medium.com/@kylejones_47003/patterns-in-time-series-for-forecasting-8a0d3ad3b7f5

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Simulated beehive data
date_range = pd.date_range(start="2024-01-01", periods=48, freq="H")
data = {'temperature': np.random.normal(35, 2, len(date_range)),
        'weight': 50 + np.cumsum(np.random.normal(0.2, 0.1, len(date_range))),
        'bee_traffic': np.sin(np.linspace(0, 3*np.pi, len(date_range))) * 50 + 200}
df = pd.DataFrame(data, index=date_range)
df.index.name = 'time'
df.head()

# Naive Forecast
df['naive_weight'] = df['weight'].shift(1)
# Moving Average Forecast
df['sma_weight'] = df['weight'].rolling(window=6).mean()
# Plotting forecasts
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['weight'], label="Actual Weight", color='brown')
plt.plot(df.index, df['naive_weight'], label="Naive Forecast", linestyle='-')
plt.plot(df.index, df['sma_weight'], label="6-Hour SMA Forecast", linestyle='--', color='red')
plt.title("Forecasting Hive Weight")
plt.xlabel("Time")
plt.ylabel("Hive Weight (kg)")
plt.legend()
plt.savefig("beehive_forecast.png")
plt.show()
```

# imputing missing values

## Forward fill

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Example Time Series with Missing Values
date_range = pd.date_range(start='2023-01-01', periods=10, freq='D')
data = [10, 12, np.nan, np.nan, 15, np.nan, 18, 20, np.nan, 22]
df = pd.Series(data, index=date_range)

# Forward Fill
df_ffill = df.ffill()

# Plot Original and Forward-Filled Data
plt.figure(figsize=(12, 6))
plt.plot(df.index, df.values, label='Original', marker='o')
plt.plot(df_ffill.index, df_ffill.values, label='Forward Fill', marker='x')
plt.legend()
plt.title("Forward Fill for Missing Values")
plt.xlabel("Date")
plt.ylabel("Value")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('forward_fill_missing_values.png')
plt.show()
```

## Backward fill

```python
# Backward Fill
df_bfill = df.bfill()

# Plot Original and Forward-Filled Data
plt.figure(figsize=(12, 6))
plt.plot(df.index, df.values, label='Original', marker='o')
plt.plot(df_bfill.index, df_bfill.values, label='Back Fill', marker='x')
plt.legend()
plt.title("Forward Fill for Missing Values")
plt.xlabel("Date")
plt.ylabel("Value")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('back_fill_missing_values.png')
plt.show()
```

## Mean fill

```python
# Forward Fill
df_mfill = df.fillna(df.mean())

# Plot Original and Forward-Filled Data
plt.figure(figsize=(12, 6))
plt.plot(df.index, df.values, label='Original', marker='o')
plt.plot(df_mfill.index, df_mfill.values, label='Mean Fill', marker='x')
plt.legend()
plt.title("Mean Fill for Missing Values")
plt.xlabel("Date")
plt.ylabel("Value")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('mean_fill_missing_values.png')
plt.show()
```

## Regression based imputation

```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Prepare Data
data_with_gaps = pd.DataFrame({'Value': data})
data_with_gaps['Index'] = np.arange(len(data))
train_data = data_with_gaps.dropna()

# Train Model
model = LinearRegression().fit(train_data[['Index']], train_data['Value'])

# Impute Missing Values
data_with_gaps.loc[data_with_gaps['Value'].isnull(), 'Value'] = model.predict(
    data_with_gaps.loc[data_with_gaps['Value'].isnull(), ['Index']]
)

# Plot Results
plt.plot(date_range, data, 'o-', label='Original')
plt.plot(date_range, data_with_gaps['Value'], 'x-', label='Imputed')
plt.legend()
plt.xticks(rotation=45)
plt.savefig('regression_fill_missing_values.png')
plt.show()
```

# Feature Engineering


## Scaling values

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import numpy as np
# Example Time Series
y = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
# Min-Max Scaling
scaler = MinMaxScaler()
y_minmax = scaler.fit_transform(y)
# Standardization
scaler_std = StandardScaler()
y_std = scaler_std.fit_transform(y)
print("Min-Max Scaled:", y_minmax)
print("Standardized:", y_std.flatten())
```

## Differencing

```python
import pandas as pd
# Example Time Series
y = pd.Series([10, 12, 15, 19, 24])
# First Difference
y_diff = y.diff().dropna()
print("First Difference:", y_diff.values)
```

## Derivatives

```python
import numpy as np
# Example Time Series
y = np.array([10, 12, 15, 19, 24])
# First Derivative (Rate of Change)
dy = np.gradient(y)
print("First Derivative (Rate of Change):", dy)
# Second Derivative (Acceleration)
d2y = np.gradient(dy)
print("Second Derivative (Acceleration):", d2y)
```

# Visualization

## Line Plot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Time Series Data
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url, parse_dates=['Month'], index_col='Month')

# Plot the Time Series
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Passengers'], color='blue', label='Passengers')
plt.title("Monthly Airline Passenger Data")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.tight_layout()
plt.savefig('airline_passengers.png')
plt.show()

print(df.head())
print(f"Data range: {df.index.min()} to {df.index.max()}")
```

## Downsampling Example: Convert monthly data to yearly averages.

```python
df_yearly = df.resample('Y').mean()

# Plot the Time Series
plt.figure(figsize=(10, 6))
plt.plot(df_yearly.index, df_yearly['Passengers'], color='blue', label='Passengers')
plt.title("Yearly Airline Passenger Data")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.tight_layout()
plt.savefig('airline_passengers_yearly.png')
plt.show()

print(df_yearly.head())
print(f"Data range: {df_yearly.index.min()} to {df_yearly.index.max()}")
```

## Upsampling Example: Convert monthly data to daily with interpolation.

```python
df_daily = df.resample('D').interpolate(method='cubic')

# Plot the Time Series
plt.figure(figsize=(15, 6))
plt.plot(df_daily.index, df_daily['Passengers'], color='blue', label='Passengers (Interpolated)')
plt.title("Daily Airline Passenger Data (Interpolated)")
plt.xlabel("Date")
plt.ylabel("Number of Passengers")
plt.legend()
plt.tight_layout()
plt.savefig('airline_passengers_daily.png')
plt.show()

print(df_daily.head())
print(f"Data range: {df_daily.index.min()} to {df_daily.index.max()}")
print(f"Total data points: {len(df_daily)}")
```

## Rolling Mean

```python
rolling_mean = df.rolling(window=12).mean()
# Plot Original and Rolling Mean
plt.figure(figsize=(10, 6))
plt.plot(df.index, df.values, label='Original Data', alpha=0.5)
plt.plot(rolling_mean.index, rolling_mean.values, label='12-Month Rolling Mean', color='red', linewidth=2)
plt.title("12-Month Rolling Mean of Airline Passengers")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.savefig('airline_passengers_rolling_mean.png')
plt.show()
```

## Seasonal Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Perform seasonal decomposition
decomposition = seasonal_decompose(df['Passengers'], model='additive', period=12)

# Plot the Decomposed Components
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(15, 20))

ax1.plot(df.index, df['Passengers'], label='Original')
ax1.set_title('Original Time Series')
ax1.legend()

ax2.plot(decomposition.trend, label='Trend', color='red')
ax2.set_title('Trend')
ax2.legend()

ax3.plot(decomposition.seasonal, label='Seasonality', color='green')
ax3.set_title('Seasonality')
ax3.legend()

ax4.plot(decomposition.resid, label='Residuals', color='orange')
ax4.set_title('Residuals')
ax4.legend()

plt.tight_layout()
plt.savefig('airline_passengers_decomposition.png')
plt.show()
```

##  Heatmap of Seasonal Data

```python
import seaborn as sns

# Reshape Data for Heatmap (Year-Month)
df = df.reset_index()
df['Year'] = df['Month'].dt.year
df['Month'] = df['Month'].dt.month
pivot_table = df.pivot_table(index='Year', columns='Month', values='Passengers')

# Plot Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='coolwarm', annot=True, fmt=".0f")
plt.title("Heatmap of Monthly Airline Passengers")
plt.xlabel("Month")
plt.ylabel("Year")
plt.tight_layout()
plt.savefig('airline_passengers_heatmap.png')
plt.show()
```

## Highlighting Anomalies

```python
window = 12  # 12-month rolling window
rolling_mean = df['Passengers'].rolling(window=window).mean()
rolling_std = df['Passengers'].rolling(window=window).std()

# Calculate dynamic threshold
threshold = rolling_mean + 2 * rolling_std

# Identify anomalies
anomalies = df[df['Passengers'] > threshold]

# Plot
plt.figure(figsize=(15, 8))
plt.plot(df.index, df['Passengers'], label="Original Data")
plt.plot(threshold.index, threshold, color='green', linestyle='--', label='Dynamic Threshold')
plt.scatter(anomalies.index, anomalies['Passengers'], color='red', s=50, label="Anomalies")

plt.title("Anomaly Detection in Airline Passengers (Dynamic Threshold)")
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.legend()

plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.savefig('airline_passengers_dynamic_anomalies.png')
plt.show()

print("\nDetected Anomalies:")
print(anomalies)
```
