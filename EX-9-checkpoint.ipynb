```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report as cr
from sklearn.model_selection import train_test_split
data = pd.read_csv('Monthly_subspots.csv', parse_dates = True, index_col = 0).squeeze()
```

```python
#ACF and PACF
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(data, lags=70)
plt.title('Autocorrelation Function (ACF)')
plt.show()
plot_pacf(data, lags=50)
plt.title('Partial Autocorrelation Function (PACF)')
plt.show()
```

```python
from pandas.plotting import autocorrelation_plot
autocorrelation_plot(data)
plt.title('Autocorrelation Plot of Shampoo Sales')
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(data, order=(5,1,0))
model_fit = model.fit()

# Summary of the fit model
print(model_fit.summary())
```

```python
residuals = pd.DataFrame(model_fit.resid)
residuals.plot()
plt.title('Residuals Line Plot')
plt.show()
```

```python
residuals.plot(kind='kde')
plt.title('Residuals Density Plot')
plt.show()
```

```python
residuals.describe()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
X = data.values
size = int(len(X) * 0.66)
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]
predictions = list()

# Walk-forward validation
for t in range(len(test)):
    model = ARIMA(history, order=(5,1,0))
    model_fit = model.fit()
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)
    print(f'predicted={yhat}, expected={obs}')
```

```python
rmse = sqrt(mean_squared_error(test, predictions))
print(f'Test RMSE: {rmse:.3f}')
```

```python
#plot expected and predicted
plt.plot(test, label='Expected')
plt.plot(predictions, color='red', label='Predicted')
plt.title('Shampoo Sales: Expected vs Predicted')
plt.xlabel('Time')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

# Plot of ARMA, ARIMA, IMA

```python
import warnings
warnings.filterwarnings('ignore')

arma_mod20 = ARIMA(data, order=(2, 0, 1)).fit()
print("ARMA(2,0,1) Parameters:\n", arma_mod20.params)
# Get the fitted values (predictions from the model)
fitted_values = arma_mod20.fittedvalues

# Plot the actual data vs the fitted values
plt.figure(figsize=(10, 6))
plt.plot(data, label='Actual Data', color='brown')
plt.plot(fitted_values, label='Fitted Values', color='pink')
plt.title('ARMA(2,0,1) Model: Actual vs Fitted Values')
plt.xlabel('Time')
plt.ylabel('Values')
plt.legend()
plt.show()
```
