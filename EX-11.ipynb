```python
import pandas as pd

custom_column_names = ['WSR0', 'WSR1']

# Read without parsing dates first
df = pd.read_csv("eighthr.data.csv", header=None, na_values=['?'], names=custom_column_names + list(range(3, 74)))

# Convert 'WSR0' to datetime if needed
# df['WSR0'] = pd.to_datetime(df['WSR0'], format='%Y-%m-%d %H:%M:%S')
```

```python
import matplotlib.pyplot as plt

# Create Figure and Axes
fig, axes = plt.subplots(2, 1, figsize=(14, 8))  # Correct way

# Plot data
df['WSR1'].plot(ax=axes[0], title='WSR0')  
df['WSR0'].plot(ax=axes[1], title='WSR1') 

plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(df['WSR0'])
print(result)
if result[1]<0.05:
    print("it is stationary")
else:
    print("it is non stationary")
```

```python
result = adfuller(df['WSR1'])
print(result)
if result[1]<0.05:
    print("it is stationary")
else:
    print("it is non stationary")
```

```python
from statsmodels.tsa.stattools import grangercausalitytests
from statsmodels.tsa.statespace.varmax import VARMAX
```

```python
print('WSR0 causes WSR1')
print('--------------------')
granger1=grangercausalitytests(df[['WSR0','WSR1']],2)
print('WSR1 causes WSR0')
print('--------------------')
granger1=grangercausalitytests(df[['WSR1','WSR0']],2)
```

```python
import pandas as pd
from statsmodels.tsa.api import VAR
from sklearn.model_selection import train_test_split


train, test = train_test_split(df[['WSR0', 'WSR1']], test_size=0.2, shuffle=False)

# Create and fit the VAR model
model = VAR(train)
sortedmodel = model.select_order(maxlags=20)
print(sortedmodel.summary())
```

```python
var_model= VARMAX(train,order=(10,0),enforce_stationary=True)
fitmodel= var_model.fit()
fitmodel.summary()
```

```python
n_forecast=12
pred=fitmodel.get_prediction()
preds=pred.predicted_mean
```

```python
preds.columns=['WSR0 Predictions','WSR1 Predictions']
preds
```

```python
testvspred= pd.concat([train,preds],axis=1)
testvspred
```

```python
testvspred[['WSR0','WSR0 Predictions']].plot()
```

```python
testvspred[['WSR1','WSR1 Predictions']].plot()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(testvspred['WSR1'],testvspred['WSR1 Predictions'])
```
